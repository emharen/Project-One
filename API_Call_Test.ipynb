```python
import gmaps
import pandas as pd
import numpy as np
import requests
import json

from config import gkey
from config2 import sodakey

# Configure gmaps
gmaps.configure(api_key=gkey)
gmaps.figure()
```

```python
base_url = "https://data.cityofchicago.org/resource/5qvk-gt8i.json"

headers = {
    "key": sodakey
}

response = requests.get(base_url, headers=headers).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
go_live = []
lat = []
lng = []
address = []
i = 0

while i < len(response):
    go_live.append(response[i]['go_live_date'])
    lat.append(float(response[i]['latitude']))
    lng.append(float(response[i]['longitude']))
    address.append(response[i]['address'])
    i += 1

speed_cam = pd.DataFrame({"Go-Live Date": go_live,
                         "Lat": lat,
                         "Lng": lng,
                         "Address": address})

speed_cam.head()
```

```python
speed_loc = speed_cam.loc[:,"Lat":"Lng"]
speed_loc.head()
```

```python
address_list = speed_cam.loc[:,"Address"]

speed_address_layer = gmaps.symbol_layer(
    speed_loc, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=[f"Location: {address}" for address in address_list]
)

fig = gmaps.figure()

fig.add_layer(speed_address_layer)

fig 
```
