```python
import pandas as pd
```

```python
file = "Traffic_Crashes_-_Crashes.csv"

crashes_df = pd.read_csv(file)
```

```python
crashes_df.head()
```

```python
crashes_df.describe()
```

```python
#Separate data by years:
crashes_df["Year"] = ""

#Split column for crashdate on date
crashes_df["Year"] = pd.DatetimeIndex(crashes_df["CRASH_DATE"]).year
```

```python
#Create a dataframe for total accidents per year
yearly = crashes_df.groupby(["Year"])

years = yearly["Year"].mean()
accidentsperyear = yearly["RD_NO"].count()
injuriesperyear = yearly["INJURIES_TOTAL"].sum()
fatalitiesperyear = yearly["INJURIES_FATAL"].sum()

yearlysummary_df = pd.DataFrame({"Accidents" : accidentsperyear, "Injuries" : injuriesperyear, "Fatalities" : fatalitiesperyear})
yearlysummary_df
```

```python
#Find all the accidents reported to have occured in intersections with stoplights that were functioning at the time
atstoplight = crashes_df.loc[crashes_df['TRAFFIC_CONTROL_DEVICE'] == 'TRAFFIC SIGNAL', :]
atstoplight = atstoplight.loc[atstoplight['DEVICE_CONDITION'] == 'FUNCTIONING PROPERLY', :]
atstoplight
```
