```python
import pandas as pd
```

```python
file = "CSVs/Speed_Camera_Locations.csv"
speed_df = pd.read_csv(file)
speed_df
```

```python
file2 = "CSVs/Chicago_Traffic_Tracker_-_Congestion_Estimates_by_Regions.csv"
regions_df = pd.read_csv(file2)
regions_df
```

```python
speed_df.rename(columns={"LATITUDE": "Lat", "LONGITUDE":"Lng"}, inplace=True)
speed_df["Region ID"] = ""
speed_df["Region Name"] = ""
counter = 0
for index, row in speed_df.iterrows():
    for index2, row2 in regions_df.iterrows():
        if ((speed_df["Lat"][index] < regions_df[" NORTH"][index2]) & (speed_df["Lat"][index] > regions_df[" SOUTH"][index2]) & (speed_df["Lng"][index] > regions_df[" WEST"][index2]) & (speed_df["Lng"][index] < regions_df[" EAST"][index2])):
            speed_df["Region ID"][index] = regions_df[" REGION_ID"][index2]
            speed_df["Region Name"][index] = regions_df["REGION"][index2]
            counter += 1
            if counter % 1000 == 0:
                print(f"{counter} records matched...")
```
