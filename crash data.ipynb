```python
file = "Traffic_Crashes_-_Crashes_2017-Present.csv"
Crashes_date = pd.read_csv(file)
Crashes_date.head()
```

```python
Crashes_date["Date"]=""
```

```python
counter=0
for index,row in Crashes_date.iterrows():
    crash_date= Crashes_date["CRASH_DATE"][index]
    Crashes_date["Date"][index]= crash_date.split()[0]
    counter+=1
    if counter %1000==0:
        print(counter)
Crashes_date.head()
```
