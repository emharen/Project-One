```python
import pandas as pd
```

```python
crashesfile = "crashesrevised.csv"
peoplefile = "peoplerevised.csv"
vehiclesfile = "vehiclesrevised.csv"
redlightfile = "redlightrevised.csv"
speedfile = "speedrevised.csv"

crashes_df = pd.read_csv(crashesfile)
people_df = pd.read_csv(peoplefile)
vehicles_df = pd.read_csv(vehiclesfile)
redlight_df = pd.read_csv(redlightfile)
speed_df = pd.read_csv(speedfile)
```

```python
speed_df.columns
```

```python
#Get rid of zero values in latitude and longitude
crashes_df = crashes_df[crashes_df.LATITUDE != 0]
crashes_df = crashes_df[crashes_df.LONGITUDE != 0]

speed_df = speed_df[speed_df.LATITUDE != 0]
speed_df = speed_df[speed_df.LONGITUDE != 0]

redlight_df = redlight_df[redlight_df.LATITUDE != 0]
redlight_df = redlight_df[redlight_df.LONGITUDE != 0]
```

```python
file = "Chicago_Traffic_Tracker_-_Congestion_Estimates_by_Regions.csv"
regions_df = pd.read_csv(file)
regions_df.head()
```

```python
crashes_df["Region"] = ""
crashes_df["Region ID"] = ""
counter = 0

for index, row in crashes_df.iterrows():
    for index2, row2 in regions_df.iterrows():
        if ((crashes_df["LATITUDE"][index] < regions_df[" NORTH"][index2]) & (crashes_df["LATITUDE"][index] > regions_df[" SOUTH"][index2]) & (crashes_df["LONGITUDE"][index] < regions_df[" WEST"][index2]) & (crashes_df["LONGITUDE"][index] > regions_df[" EAST"][index2])):
            counter += 1
            crashes_df["Region ID"][index] = regions_df[" REGION ID"][index2]
            crashes_df["Region"][index] = regions_df["REGION"][index2]
        if counter % 1000 == 0:
            print(str(counter) + "records matched.")
        
            
       
            
```
