```python
file = "Chicago_Traffic_Tracker_-_Congestion_Estimates_by_Regions.csv"
regions_df = pd.read_csv(file)
regions_df.head()
```

```python
crashes_df["Region"] = ""
crashes_df["Region ID"] = ""

for index, row in crashes_df.iterrows():
    for index2, row2 in regions_df.iterrows():
        if (row["LATITUDE"] < row2[" NORTH"]) & (row["LATITUDE"] > row2[" SOUTH"]) & (row["LONGITUDE"] < row2[" WEST"]) & (row["LONGITUDE"] > row2[" EAST"]):
            crashes_df["Region ID"] = row2[" REGION ID"]
            crashes_df["Region"] = row2["REGION"]
            
```
