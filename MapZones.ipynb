```python
import pandas as pd
import gmaps
from googleapikey import gkey
```

```python
file = "Chicago_Traffic_Tracker_-_Congestion_Estimates_by_Regions.csv"
zones_df = pd.read_csv(file)
zones_df.head()
```

```python
zones_df[["REGION", " REGION_ID"," WEST", " EAST", " SOUTH", " NORTH"]]
```

```python
gmaps.configure(api_key = gkey)
gmaps.figure()
```

```python
#Create a list of coordinates for our figure
SW = []
NW = []
SE = []
NE = []

for index, row in zones_df.iterrows():
    temp = []
    temp.append(row[" NORTH"])
    temp.append(row[" EAST"])
    NE.append(temp)

for index,row in zones_df.iterrows():
    temp = []
    temp.append(row[" NORTH"])
    temp.append(row[" WEST"])
    NW.append(temp)
    
for index,row in zones_df.iterrows():
    temp = []
    temp.append(row[" SOUTH"])
    temp.append(row[" EAST"])
    SE.append(temp)
    
for index,row in zones_df.iterrows():
    temp = []
    temp.append(row[" SOUTH"])
    temp.append(row[" WEST"])
    SW.append(temp)    

print(SW)
```

```python
north = zones_df[" NORTH"]
south = zones_df[" SOUTH"]
west = zones_df[" WEST"]
east = zones_df[" EAST"]
    
    
print(north)    
```

```python
#Make coordinates into a distinct zone list
coords = []

for x in range(29):
    temp = []
    temp.append(NE[x])
    temp.append(SE[x])
    temp.append(SW[x])
    temp.append(NW[x])
    coords.append(temp)
    
print(coords[0])
```

```python
#Make the map
fig = gmaps.figure(center = (41.835966, -87.686252), zoom_level = 11)
#drawing = gmaps.drawing_layer(features = 
                             #gmaps.Polygon(north[0], west[0], north[0], east[0], south[0], west[0], south[0], east[0]))

for x in range(29):
    region_layer = gmaps.drawing_layer(features = [
        gmaps.Polygon(coords[x],
                     stroke_color = 'blue', stroke_weight = 1, fill_color = 'blue', fill_opacity = .2)
    ])
    fig.add_layer(region_layer)
fig
```
