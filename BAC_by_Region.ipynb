```python
crashes = pd.read_csv("Resources/Traffic_Crashes_-_Crashes_2017-present.csv", low_memory=False)
crashes.head()
```

```python
merged = pd.merge(people,crashes,on='RD_NO')
merged.head()
```

```python
merged_clean = merged.loc[:,['CRASH_DATE_x','BAC_RESULT VALUE','LATITUDE','LONGITUDE']]
merged_clean.head()
```

```python
merged_clean.describe()
```

```python
merged_clean2 = merged_clean.loc[merged_clean['BAC_RESULT VALUE'] > .08,:]
merged_clean2.describe()
```

```python
merged_clean3 = merged_clean2.loc[merged_clean2['BAC_RESULT VALUE'] <= .4,:]
merged_clean3.describe()
```

```python
merged_clean3.rename(columns={'LATITUDE':'Lat','LONGITUDE':'Lng'},inplace=True)
merged_clean3['Region_ID'] = ""
merged_clean3.describe()
```

```python
regions = pd.read_csv("Resources/Chicago_Traffic_Tracker_-_Congestion_Estimates_by_Regions.csv", low_memory=False)
regions.head()
regions.describe()
```

```python
region_location = regions.loc[:," REGION_ID":" NORTH"]
region_location.head()
```

```python
for index, row in merged_clean3.iterrows():
    for index2, row2 in region_location.iterrows():
        if ((merged_clean3["Lat"][index] < region_location[" NORTH"][index2]) & (merged_clean3["Lat"][index] > region_location[" SOUTH"][index2]) & (merged_clean3["Lng"][index] > region_location[" WEST"][index2]) & (merged_clean3["Lng"][index] < region_location[" EAST"][index2])):
            merged_clean3["Region_ID"][index] = region_location[" REGION_ID"][index2]
merged_clean3.head()
```

```python
merged_clean3.describe()
```

```python
merged_clean3.drop(merged_clean3[merged_clean3.Region_ID == ""].index, inplace=True)
# df.drop(df[df.score < 50].index, inplace=True)
merged_clean3.describe()

```

```python
merged_clean4 = merged_clean3.loc[:,['Region_ID','BAC_RESULT VALUE']]
merged_clean4.head()
```

```python
region_bac = merged_clean4.groupby('Region_ID')
region_bac.count()
```

```python
region_bac_df = pd.DataFrame(region_bac.count())
region_bac_df.head()
```

```python
region_bac_df.to_csv('Resources/region_bac_count.csv')
```
